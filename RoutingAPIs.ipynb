# Introduction 

__The Problem of Routing__

Routing or navigation for a clearer sense of the term, is the problem of finding the most convenient way to get one one place on the map to the other.


__Measure of Convenience__ : 
Convenience here is a bread term for the kind of conditions we want to be able to provide to find the most suitable route for ourselves.
For example:
1. shortest distance
2. least traffic
3. An ATM on the route, etc


# Scope of the Notebook 

The problem that we are considering at the moment is shortest distance problem.

Note: we're calculating distance and not displacement.

Imagine an algorithm that works on obstacle detection principle that is follows the path of displacement, until an obstacle is met, in the case of which it re-routes. Such a path finding technique would not only be inaccurate and cumbersome but also very expensive in real life scenario, where there can be many hindrances.
Hence the need for a Routing Algorithm that ananlyses all possible routes from the source to the destination and then takes an informed call on which has the least cost.

# About A* Algorithm

The function is a combination of two elemental functions, g(n) : which is the distance between the the source and node n and h(n): A heuristic function estimating the distance between node n and the destination.

Based on the combined result of these two parameters, the algorithm assigns a cost to each path.
The heuristic function is what gives a sense of direction to the A* Algortihm which Djikstra lacks.

f(n) = g(n) + h(n)

_Advantages of using A* Algorithm_:
1. Complete and Optimal 
2. Efficient and covers the minimum no. of nodes required to cover.

_Disadvantages od using A* Algorithm_:

1. Executables only if the braching factor is finite.
and
2. Every actio (here choice of path) has a fixed cost.

# 3. OSRM (Open Street Routing Machine)

__Data__ : OpenStreetMap <br>

In contrast to A*. This inturn makes it much faster than the above APIs. Being BSD liscenced it is highly convenient to use OSRM.
